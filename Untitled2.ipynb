```python
# outlier
def Handling_Outliers(df):
    for col in df.select_dtypes(include = ['float64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        Lower_bound = Q1 - (1.5 * IQR)
        Upper_bound = Q3 + (1.5 * IQR)

        #CAPPING OUTLIERS
        df[col]= df[col].apply(lambda x: Lower_bound if x < Lower_bound else Upper_bound if x > Upper_bound else x)

    return df
```

```python
df1 = Handling_Outliers(df)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
X = df1.drop(['Natural resource depletion (% of GNI) 2008Ð2013'],axis = 1)
y = df1['Natural resource depletion (% of GNI) 2008Ð2013']
```

```python
X = df1.drop(['Population affected by Natural disasters (average annual per million people) 2005/2014'], axis = 1)
y = df1['Population affected by Natural disasters (average annual per million people) 2005/2014']
```

```python
sk = SelectKBest(score_func=f_regression,k=10)
x_new = sk.fit_transform(X,y)
selected_features = X.columns[sk.get_support()]
feature_score = pd.DataFrame({'feature':X.columns,'Score':sk.scores_}).sort_values(by='Score',ascending=False)
print("selected_Fetaures:",selected_features)
print("\nFeature Score:")
print(feature_score)
```

```python
X_selected = X[selected_features]
```

```python
X_selected.columns
```

```python
# SCALING
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X_selected)
x_scaled
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV, GridSearchCV
```

```python
#splitting
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2,random_state =42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
print(f""" 
X_train shape : {X_train.shape}
X_test shape : {X_test.shape}
y_train shape : {y_train.shape}
y_test shape : {y_test.shape} 
""")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
models ={
    '1.Linear Regression' : LinearRegression(),
    '2.Decision Tree Regressor' : DecisionTreeRegressor(),
    '3.Random Forest Regressor': RandomForestRegressor(),
    '4.Gradient Boosting Regressor': GradientBoostingRegressor(),
    '5.SVR' : SVR()
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
result={}
for model_name,model in models.items():
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    MAE = mean_absolute_error(y_test,y_pred)
    MSE = mean_squared_error(y_test,y_pred)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test,y_pred)
    result[model_name]={'MAE':MAE,'MSE':MSE,'RMSE':RMSE,'R2':R2}
```

```python
result_df = pd.DataFrame(result).T
```

```python
result_df
```

```python
#Hyperparameter tuning
rfg = RandomForestRegressor(
    random_state=42,
    n_estimators=50,         
    max_depth=10,             
    min_samples_split=5,      
    min_samples_leaf=2,       
    max_features='sqrt',      
    n_jobs=-1                
)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200,300],  
    'max_depth': [10, 20, None],     
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4]   
}
```

```python
grid_search = GridSearchCV(
    estimator=rfg,
    param_grid=param_grid,
    cv=10,  
    scoring='r2',  
    n_jobs=-1, 
    verbose=2   
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python

y_pred = best_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test,y_pred)
```

```python
print(f" MAE :{MAE}, MSE: {MSE}, RMSE :{RMSE}, R2:{R2}") 
```
